```python
# Importing libraries and dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
dataset = fetch_openml("diabetes" , version = 1 , as_frame = True)
x = pd.DataFrame(dataset.data)
y = pd.DataFrame(dataset.target)
x.head()
```

```python
# EDA - Identifying patterns and missing data

# No missing data
x.isnull().sum()

# Every columns has Numeric data , no need of Encoding
x.head()

# Target contains data in string datatype , encoding y into numeric data
y["class"].replace({'tested_negative' : 0 , "tested_positive" : 1} , inplace = True)
print(y.head())
```

```python
# Dividing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# model importing and training


# from sklearn.linear_model import LogisticRegression
# model = LogisticRegression(max_iter=100) # Accuracy = 72~74

# from sklearn.naive_bayes import GaussianNB
# model = GaussianNB() #Accuracy = ~74

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier() # = 75~80

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

from sklearn import metrics
acc = metrics.accuracy_score(y_pred,y_test)
print("Accuracy : " , acc)
cm = metrics.confusion_matrix(y_pred,y_test)
print("Confusion matrix:\n" , cm)
```
